```python
import pandas as pd

# We have done for each subgroups
file_list = [
    '5ci_qrup_2021_2022.csv',
    '5ci_qrup_2022_2023.csv',
    '5ci_qrup_2023_2024.csv',
    '5ci_qrup_2024_2025.csv',
    '5ci_qrup_2025_2026.csv'
]


df_list = []
for file in file_list:
   
    df = pd.read_csv(file, sep=';', skiprows=1)
    df_list.append(df)


merged_df = pd.concat(df_list, ignore_index=True)


merged_df['year'] = merged_df['year'].fillna('2021-2022')
merged_df['year'] = merged_df['year'].astype(str)


output_filename = 'merged_5ci_qrup_corrected.csv'
merged_df.to_csv(output_filename, sep=';', index=False)

print(f"Successfully merged files and saved to: {output_filename}")
```
